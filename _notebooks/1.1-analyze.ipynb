```python
df["label"] = df["label"].replace("[]", "")

df.head(3)
```

```python
df["label"] = df["label"].str.replace("['\[\]]", "", regex=True)

labels = df["label"].to_list()

labels = [x.split(",") for x in labels]

labels = [item for sublist in labels for item in sublist]

labels = [x.lstrip() for x in labels]

labels_count = pd.Series(labels).value_counts()

labels_count
```

```python
labels = [
    "women",
    "politics",
    "appearance",
    "disabled",
    "racism",
    "class",
    "lgbti",
    "criminal",
    "calls",
]

for label in labels:
    df[label] = df["label"].str.contains(label, case=False).astype(int)

df.head(3)
```

```python
df["to_journalist"] = (
    df["query_type"]
    .str.extract(r"(?<=to:\s)(@\w+)", flags=re.IGNORECASE)
    .fillna(np.nan)
)

df.head(3)
```

```python
df["from_journalist"] = (
    df["query_type"]
    .str.extract(r"(?<=from:\s)(@\w+)", flags=re.IGNORECASE)
    .fillna(np.nan)
)

df.head(3)
```

```python
df["label"] = df["label"].replace("", np.nan)

attacks = df.dropna(subset=["label", "to_journalist"])
attacks.head(3)
```

```python
df.to_csv("../data/processed/elecciones_argentina.csv", index=False)
attacks.to_excel(
    "../data/processed/ataques_elecciones_argentina.xlsx",
    index=False,
    sheet_name="data",
)
```
