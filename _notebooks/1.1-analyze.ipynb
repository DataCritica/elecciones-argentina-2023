```python
df['journalist_username'] = df['query_type'].str.extract(r'(@\w+)')
df.head(3)
```

```python
mujeres = ['@Miglesias71', '@gabipellegrini3', '@GabrielaWeller',
            '@SanchezClarise', '@soyingridbeck', '@SilvinaMolina',
            '@mafito11', '@AJuliaLopez', '@Cris_noticias',
            '@vivicanosaok', '@marialauratv', '@guadavazquez',
            '@odonnellmaria', '@mjolivan', '@maclorena',
            '@rominamanguel', '@monigps', '@lucianageuna',
            '@negropolisok', '@juliamengo', '@FlorHalfon',
            '@cyngarciaradio', '@deboraplager', '@silviafbarrio',
            '@NANCYPAZOS', '@gabycociffi', '@SoleVallejos',
            '@nbg__', '@hindelita', '@MercedesFunes',
            '@anaecorrea', '@Marcelitaojeda', '@lucianapeker',
            '@Angelalerena']

hombres = ['@rialjorge', '@fantinofantino', '@andykusnetzoff',
           '@SANTIAGODELMORO', '@VHMok', '@wwnicolas',
           '@edufeiok', '@luisnovaresio', '@JonatanViale',
           '@majulluis', '@barilirodolfo', '@diegoleuco',
           '@alfleuco', '@Gatosylvestre', '@robdnavarro',
           '@Sietecase', '@aleberco', '@diegobranca',
           '@ischargro', '@ertenembaum']

print(f'Mujeres periodistas: {len(mujeres)}\nHombres periodistas: {len(hombres)}')

# Check if any element from the lists is present in the 'query_type' column
df['journalist_genre'] = ''
df.loc[df['query_type'].str.contains('|'.join(mujeres)), 'journalist_genre'] = 'M'
df.loc[df['query_type'].str.contains('|'.join(hombres)), 'journalist_genre'] = 'H'

df.head(3)
```

```python
df['label'] = df['label'].replace("[]", '')

df.head(3)
```

```python
df['label'] = df['label'].str.replace("[\'\[\]]", '', regex=True)

labels = df['label'].to_list()

labels = [x.split(',') for x in labels]

labels = [item for sublist in labels for item in sublist]

labels = [x.lstrip() for x in labels]

labels_count = pd.Series(labels).value_counts()

labels_count
```

```python
labels = ['women', 'politics', 'appearance', 'disabled', 'racism', 'class', 'lgbti', 'criminal', 'calls']

for label in labels:
    df[label] = df['label'].str.contains(label, case=False).astype(int)

df.head(3)
```

```python
df['to_journalist'] = df['query_type'].str.extract(r'(?<=to:\s)(@\w+)', flags=re.IGNORECASE).fillna(np.nan)

df.head(3)
```

```python
df['from_journalist'] = df['query_type'].str.extract(r'(?<=from:\s)(@\w+)', flags=re.IGNORECASE).fillna(np.nan)

df.head(3)
```

```python
attacks = df.dropna(subset=['label', 'to_journalist'])

attacks.head(3)
```

```python
df.to_csv('../data/processed/elecciones_argentina.csv', index=False)
attacks.to_excel('../data/processed/ataques_elecciones_argentina.xlsx', index=False, sheet_name='data')
```
